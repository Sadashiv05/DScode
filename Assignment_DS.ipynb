```python
#Renaming few columns
student_data.rename(columns={'Python (out of 3)': 'Python','R Programming (out of 3)':'R','Data Science (out of 3)':'Data Science'}, inplace=True)
```

```python
# Create a dataframe to store the result
result_data = pd.DataFrame(columns=['app_id','total_score'])
```

```python
for row_num, content in student_data.iterrows():
    score=0
    
    if ('B.E' in str(content['Degree']) or 'B.Tech' in str(content['Degree']) or 'M.Sc' in str(content['Degree']) or 'M.Tech' in str(content['Degree'])) :
        if content['Current Year Of Graduation'] == 2020 and pd.isna(content['Performance_PG']) and not pd.isna(content['Performance_UG']):
            score = 10
        elif content['Current Year Of Graduation'] == 2019 and pd.isna(content['Performance_PG']) and not pd.isna(content['Performance_UG']): 
            score = 8    
        elif content['Current Year Of Graduation'] <= 2018 and pd.isna(content['Performance_PG']) and not pd.isna(content['Performance_UG']): 
            score = 5
        elif content['Current Year Of Graduation'] == 2020 and not pd.isna(content['Performance_PG']): 
            score = 7    
        elif content['Current Year Of Graduation'] <= 2019 and not pd.isna(content['Performance_PG']) : 
            score = 3

        if content['Python'] == 3 :
            score = score + 10
        elif content['Python'] == 2 :
            score = score + 7    
        elif content['Python'] == 1 :
            score = score + 3

        if content['R'] == 3 :
            score = score + 10
        elif content['R'] == 2 :
            score = score + 7    
        elif content['R'] == 1 :
            score = score + 3

        if content['Data Science'] == 3 :
            score = score + 10
        elif content['Data Science'] == 2 :
            score = score + 7    
        elif content['Data Science'] == 1 :
            score = score + 3

        if str(content['Other skills']).find('Machine Learning') != -1  :
            score = score + 3
        if str(content['Other skills']).find('Deep Learning') != -1  :
            score = score + 3
        if str(content['Other skills']).find('NLP') != -1  :
            score = score + 3
        if str(content['Other skills']).find('Statistical Data Analysis') != -1  :
            score = score + 3
        if str(content['Other skills']).find('AWS') != -1  :
            score = score + 3
        if re.search(r'\bSQL\b',str(content['Other skills'])) or str(content['Other skills']).find('NoSQL') != -1 :
            score = score + 3
        if str(content['Other skills']).find('Excel') != -1  :
            score = score + 3

        if score >= 40 :
            result_data = result_data.append({'app_id':content['Application_ID'],'total_score':score}, ignore_index=True)  
```

```python
# Display the result : List of application id's and total score
result_data
```

```python
# Store the result in a CSV
result_data.to_csv('result.csv', index=False)
```
